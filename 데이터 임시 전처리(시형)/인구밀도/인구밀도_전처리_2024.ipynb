```python
import pandas as pd

df = pd.read_csv('./행정구역_시군구_별_주민등록세대수2023_2024.csv',encoding="cp949")
# df.info()
df.head()
```

```python
df = df.transpose()   #데이터 행과 열 바꾸기  (df.T 라고 적어도 된다!)
df   # 행과 열을 바꾸니 header에 이상한 숫자가 들어옴
```

```python
# 맨위 header에 이상한 값 제거하기 위해서 이렇게 작성하였음

df.to_csv('test.csv')
df = pd.read_csv('test.csv', header=1)
# df
df.to_csv('test.csv')
```

```python
df = df.rename(columns={'행정구역(시군구)별':'기간'})
df = df.rename(columns={'서울특별시':'인구'})
```

```python
df.isnull().sum()        #결측값 제로
```

```python
df = df.iloc[:, :3]
df = df.drop('전국', axis=1)
```
