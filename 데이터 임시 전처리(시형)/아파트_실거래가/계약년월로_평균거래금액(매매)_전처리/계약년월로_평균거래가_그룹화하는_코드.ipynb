계약년월로_평균거래가_그룹화하는_코드

```python
import pandas as pd

df_2023 = pd.read_csv('../부동산_실거래가_2023.csv', header=0, encoding='cp949')
df_2024 = pd.read_csv('../부동산_실거래가_2024.csv', header=0, encoding='cp949')

df_2023.head(2)
```

```python
df_2024.head(2)
```

```python
df=pd.concat([df_2023, df_2024])
df.head(2)
```

```python
# df.info()
df= df[['계약일', '물건금액(만원)']]
df
```

```python
# 열이름 바꾸기
df.columns = df.columns.str.replace('(','_') 
df.columns = df.columns.str.replace(')','_') 
df = df.rename(columns={'접수연도':'기간'})
df = df.rename(columns={'물건금액_만원_':'실거래가'})
```

```python
df
```

```python
df.isnull().sum()    # 결측치 없음
```

```python
df[(abs((df['실거래가']-df['실거래가'].mean())/df['실거래가'].std()))>1.96]  # 이상치 3634개

# 이상치 제거
df = df[(abs((df['실거래가']-df['실거래가'].mean())/df['실거래가'].std()))<=1.96]
```

```python
# 월별로 만들어준다
df['계약일'] = round(df['계약일']/100).astype(int)    # 에러는 나는데 제대로 데이터가 바뀐다
```

```python
result_df=df
result_df = result_df.rename(columns={'계약일': '기간'})
```

```python
result_df = result_df[result_df['기간'] >= 202300]
result_df
```

```python
result_df=pd.DataFrame(result_df.groupby(['기간'])['실거래가'].mean())
```

```python
result_df.to_csv('부동산_실거래가_전처리후_2023_2024.csv')
result_df
```
